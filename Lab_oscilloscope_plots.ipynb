```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

plt.rcParams['axes.grid'] = True
```

```python
la = 0.25
tau = 1

f = np.arange(-12, 12, 1)

def X(f):
    return 1.0 * la if f == 0 else abs(np.sin(np.pi * la * tau * f) / (np.pi * f))

X = np.vectorize(X)

plt.figure(figsize=(12, 5))

plt.xlabel("$f$")
plt.stem(f, X(f))
plt.show()
```

```python
la = 0.25
tau = 1

n = 3

f = np.linspace(-12, 12, 10000)

def X(f):
    c = 0
    for i in range(n):
        c += np.exp(-2.j * np.pi * f * i * tau)
        
    # c = (np.exp(-2.j * np.pi * f * n * tau) - 1) / (np.exp(-2.j * np.pi * f * tau) - 1)
    
    return 1.0 * la if f == 0 else abs(np.sin(np.pi * la * tau * f) / (np.pi * f) * c)

X = np.vectorize(X)

plt.figure(figsize=(12, 5))

plt.xlabel("$f$")
plt.plot(f, X(f))
plt.show()
```
